```python
from scipy.integrate import odeint
import numpy as np
def model(x,t):
  y = x[0]
  dy = x[1]  
  xdot = [[],[]]
  xdot[0] = dy
  xdot[1] = -10 - 0.9*y
  return xdot
time = np.linspace(0,10,1000)
z2 = odeint(model,[2,10^(-6)],time)

# plot results
import matplotlib.pyplot as plt
plt.plot(time,z2[:,0],'g:')
plt.plot(time,z2[:,1],'k-.')
plt.legend(['u (ODEINT)','du/dt (ODEINT)'])
plt.xlabel('Time')
plt.show()
print(z2) 
```
