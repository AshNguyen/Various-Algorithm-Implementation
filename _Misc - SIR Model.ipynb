```python
import math 
import random
import matplotlib.pyplot as plt 
import numpy as np
step=1
t0=1
t=100
S0=0.6
I0=0.4
R0=0.0
alpha=0.2
lamda=0.11
i=t0
list_S=[]
list_I=[]
list_R=[]
list_S.append(S0)
list_I.append(I0)
list_R.append(R0)
while (i<t) :
    S=S0-alpha*S0*I0
    list_S.append(S)
    I=I0+alpha*S0*I0-lamda*I0
    list_I.append(I)
    R=R0+lamda*I0
    list_R.append(R)
    S0=S
    I0=I
    R0=R
    i=i+step
count=[]
x=0
while (x<t):
    count.append(x)
    x=x+1
plt.plot(count, list_S,'r-.',label="Susceptible")
plt.plot(count,list_I,'b-.',label="Infectious")
plt.plot(count,list_R,'g-.', label="Recovered")
plt.legend()
plt.show()
```
